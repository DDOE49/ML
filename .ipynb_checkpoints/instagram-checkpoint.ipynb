```
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time

import pandas as pd
from pandas import DataFrame
import numpy as np

path="D:/kdeWorking/selenium/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.implicitly_wait(3)
driver.get("https://www.instagram.com/accounts/login/?source=auth_switcher")
element = driver.find_element_by_name("username")
element.send_keys('takesin1')
element = driver.find_element_by_name("password")
element.send_keys('rhddb1004')
element.submit()
driver.implicitly_wait(5)

#기타 팝업관련 제거
driver.find_element_by_xpath("//*[@id='react-root']/section/main/div/div/div/div/button").click()
driver.implicitly_wait(5)
driver.find_element_by_xpath("/html/body/div[4]/div/div/div/div[3]/button[2]").click()

#검색창 찾기
find=driver.find_element_by_xpath("//input[@type='text']")
keyword="수원공방거리"
find.send_keys(keyword)
#해쉬태그 검색
tag=driver.find_element_by_xpath("//*[@id='react-root']/section/nav/div[2]/div/div/div[2]/div[2]/div[2]/div/a[1]/div/div/div[1]/span")
driver.implicitly_wait(5)
tag.click()

# likecnt좋아요, date날짜,  uid사용자아이디, content내용
list_likecnt=[]
list_date=[]
list_uid=[]
list_content=[]

#첫번째 게시물 클릭
driver.find_element_by_xpath("/html/body/div[1]/section/main/article/div[1]/div/div/div[1]/div[1]/a/div/div[2]").click()

date=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/div[2]/a/time").get_attribute("title")
uid=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/header/div[2]/div[1]/div[1]/a").text
try:
    likecnt=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/section[2]/div/div/button/span").text
except NoSuchElementException:
    likecnt=0

#내용 없을때 제외
try: 
    content=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/div[1]/ul/div/li/div/div/div[2]/span").text
except NoSuchElementException: 
    content=''
list_likecnt.append(likecnt)
list_date.append(date)
list_uid.append(uid)
list_content.append(content)

#오른쪽으로 사진 넘기기
driver.find_element_by_xpath("/html/body/div[4]/div[1]/div/div/a").send_keys(Keys.RIGHT)

continue_flag=0
for i in range(1,5000):
    try:
        date=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/div[2]/a/time").get_attribute("title")
    except NoSuchElementException:
        driver.implicitly_wait(5)
        if continue_flag==0:
            continue_flag=1
        else:
            print('마지막 항목' , i)
            driver.close()
        continue
        
    date=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/div[2]/a/time").get_attribute("title")
    uid=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/header/div[2]/div[1]/div[1]/a").text
    
    try:
        likecnt=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/section[2]/div/div/button/span").text
    except NoSuchElementException:
        likecnt=0
    
    #내용 없을때 제외
    try:
        content=driver.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/div[1]/ul/div/li/div/div/div[2]/span").text
    except NoSuchElementException:
        content=''
    
    #데이터 저장
    list_likecnt.append(likecnt)
    list_date.append(date)
    list_uid.append(uid)
    list_content.append(content)
    
    #다음 내용이 더 이상 없을때
    try:
        RightStats = driver.find_element_by_xpath("/html/body/div[4]/div[1]/div/div/a[2]").text
        driver.find_element_by_xpath("/html/body/div[4]/div[1]/div/div/a").send_keys(Keys.RIGHT)
        #수집텀
        time.sleep(4)
    except NoSuchElementException:
        print("끝")
        break

dataset=[list_date,list_uid,list_likecnt,list_content]
#각 리스트가 세로로 합쳐지는게 아니라 가로로 합쳐지기 때문에 변환을 해줘야한다.
final=np.transpose(dataset)

df = DataFrame(final, columns=['DATE','ID','LIKE','CONTENT'])
print(df)
#엑셀파일 저장 C:\Users\User 에 저장됨
df.to_excel('수원공방거리1.xlsx',sheet_name='Sheet1',header=True)      

driver.close()
```
