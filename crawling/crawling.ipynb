```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup

import requests
import time
import datetime

from pandas import Series, DataFrame

import os
from random import randrange

path = 'C:/kdeWorking/Project/DL/DL/crawling/chromedriver'
driver = webdriver.Chrome(path)

url = 'https://finance.daum.net/talks'
driver.get(url)
resp = requests.get(url)
time.sleep(randrange(10,40)/10)

dateTime_list = []
title_list = []
content_list = []
user_list = []
like_list = []
unlike_list = []
comment_list = []
comments_nickName_list = []
comment_dateTime_list = []

# Log In
log_in_div = driver.find_element_by_css_selector('#btnMinidaumLogin').click()
time.sleep(randrange(10,40)/10)
daum_log_in = driver.find_element_by_css_selector('.link_dlogin').click()
time.sleep(randrange(10,40)/10)

id_form = driver.find_element_by_id('id')
id_form.send_keys('yhw1994')
pw_form = driver.find_element_by_id('inputPwd')
pw_form.send_keys('ehehej419y')
time.sleep(randrange(10,40)/10)
log_in_btn = driver.find_element_by_id('loginBtn').click()

# 해당 element가 로딩 될때까지 대기
myElem = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '.line')))

# 첫번째 글 Click
tr = driver.find_element_by_class_name('line')
time.sleep(randrange(10,40)/10)
td = tr.find_element_by_tag_name('td')
time.sleep(randrange(10,40)/10)
a = td.find_element_by_tag_name('a').click()
time.sleep(randrange(10,40)/10)

# read iframe
driver.switch_to.frame("bbsFrame")

for i in range(5000):
    comments = ''
    comments_dateTime = ''
    comments_nickName = ''
    
    time.sleep(randrange(10,40)/10)
    source = driver.page_source
    soup = BeautifulSoup(source,"html.parser")
    
    # contents crawling
    dateTime = soup.select_one('.datetime').text
    title = soup.select_one('#bbsTitle').text.strip()
    content = soup.select_one('#bbsContent').text.strip()
    nickName = soup.select_one('.nick').text
    like = soup.select_one('#approCnt').text
    unlike = soup.select_one('#opposCnt').text
    
    driver.switch_to.frame("commentList") # iframe in iframe
    comment_cnt = driver.find_element_by_css_selector('.cEm').text
    
    if eval(comment_cnt) == 0:
        time.sleep(randrange(10,40)/10)
        driver.switch_to_default_content() # iframe out
        driver.switch_to.frame("bbsFrame")
    else:
        comment          = driver.find_elements_by_css_selector('.content')
        comment_nickName = driver.find_elements_by_css_selector('.nick')
        comment_dateTime = driver.find_elements_by_css_selector('.datetime')
        time.sleep(randrange(10,40)/10)
        for i in comment:
            comments += str(i.text)
            comments += '@%'
        for j in comment_nickName:
            comments_nickName += str(j.text)
            comments_nickName += '@%'
        for k in comment_dateTime:
            comments_dateTime += str(k.text)
            comments_dateTime += '@%'
        time.sleep(randrange(10,40)/10)
        driver.switch_to_default_content() # iframe out
        driver.switch_to.frame("bbsFrame")
    
    # add to list
    dateTime_list.append(dateTime)
    title_list.append(title)
    content_list.append(content)
    user_list.append(nickName)
    like_list.append(like)
    unlike_list.append(unlike)
    comment_list.append(comments)
    comments_nickName_list.append(comments_nickName)
    comment_dateTime_list.append(comments_dateTime)
    
    # move down
#     driver.switch_to_default_content() # iframe out
#     time.sleep(2)
#     driver.switch_to.frame("bbsFrame")
    time.sleep(randrange(10,40)/10)
    driver.find_element_by_class_name('goNext').click()

driver.switch_to_default_content() # iframe out
raw_data = {
    'DATE_TIME': dateTime_list,
    'TITLE': title_list,
    'CONTENT': content_list,
    'NICKNAME': user_list,
    'LIKE': like_list,
    'UNLIKE': unlike_list,
    'COMMENT': comment_list,
    'COMMENT_NICK_NAME': comments_nickName_list,
    'COMMENT_DATE_TIME': comment_dateTime_list
}

# Date And Time
now = datetime.datetime.now()
nowDateTime = now.strftime('%Y%m%d%H%M%S')

# isDir and mkDir
if os.path.exists('.\\result') is False:
    os.mkdir('.\\result')

data = DataFrame(raw_data)
data.to_csv('.\\result\\test'+nowDateTime+'.csv')

driver.close()
```
